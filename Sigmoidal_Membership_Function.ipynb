```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt

# تعریف محدوده ورودی
x = np.arange(-5, 5, 0.1)

# تعریف پارامترهای تابع عضویت سیگموئیدی
a = 1.0  # پارامتر شیب
b = 0.0  # پارامتر مکان

# تعریف تابع عضویت سیگموئیدی
membership = fuzz.sigmf(x, b, a)

# نمایش تابع عضویت سیگموئیدی
plt.figure()
plt.plot(x, membership, 'b', linewidth=1.5)
plt.title('Sigmoidal Membership Function')
plt.ylabel('Membership')
plt.xlabel('input variable (x)')
plt.grid(True)
plt.show()
```
