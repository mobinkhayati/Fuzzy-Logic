```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# تعریف مجموعه‌های فازی A و B
A = np.array([0.5, 0.7, 0.8])
B = np.array([0.6, 0.8, 1.0])

# عملیات زیر مجموعه
def is_subset(A, B):
  return np.all(A <= B)

print(is_subset(A, B))  # True

# عملیات مکمل
def complement(A):
  return 1 - A

print(complement(A))  # [0.5, 0.3, 0.2]

# عملیات اتحاد
def union(A, B):
  return np.maximum(A, B)

print(union(A, B))  # [0.5, 0.7, 1.0]

# عملیات اشتراک
def intersection(A, B):
  return np.minimum(A, B)

print(intersection(A, B))  # [0.5, 0.8, 0.8]
```

```python
# تعریف مجموعه های فازی A و B
A = np.array([0.5, 0.7, 0.8])
B = np.array([0.6, 0.8, 1.0])

# نمایش مجموعه های A و B
plt.scatter(A, np.ones(len(A)), label="A")
plt.scatter(B, np.ones(len(B)), label="B")
plt.xlabel("amount of")
plt.ylabel("Membership")
plt.legend()
plt.show()

# محاسبه زیرمجموعه
C = A <= B
plt.scatter(A, C, label="A <= B")
plt.xlabel("amount of")
plt.ylabel("Membership")
plt.legend()
plt.show()

# محاسبه مکمل
C = np.ones(len(A)) - A
plt.scatter(A, C, label="A'")
plt.xlabel("amount of")
plt.ylabel("Membership")
plt.legend()
plt.show()

# محاسبه اجتماع
C = np.maximum(A, B)
plt.scatter(A, C, label="A | B")
plt.xlabel("amount of")
plt.ylabel("Membership")
plt.legend()
plt.show()

# محاسبه اشتراک
C = np.minimum(A, B)
plt.scatter(A, C, label="A & B")
plt.xlabel("amount of")
plt.ylabel("Membership")
plt.legend()
plt.show()
```
