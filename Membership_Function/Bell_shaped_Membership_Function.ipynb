```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt

# تعریف محدوده ورودی
x = np.arange(-10, 10, 0.1)

# تعریف پارامترهای تابع عضویت شکل زنگوله
mean = 5  # میانگین
sigma = 1.0  # پارامتر انحراف معیار
width = 2.5  # عرض پیک

# تعریف تابع عضویت شکل زنگوله
membership = fuzz.gbellmf(x, width, mean, sigma)

# نمایش تابع عضویت شکل زنگوله
plt.figure()
plt.plot(x, membership, 'b', linewidth=1.5)
plt.title('Bell-shaped Membership Function')
plt.ylabel('Membership')
plt.xlabel('input variable (x)')
plt.grid(True)
plt.show()
```
